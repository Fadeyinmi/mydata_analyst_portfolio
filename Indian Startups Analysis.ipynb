```python
# 5. find number of funding per month
df['yearmonth'].value_counts()
```

```python
def year_month_group(x):
    if x >= 201501 and x < 201601:
        return '2015'
    elif x >= 201601 and x < 201701:
        return '2016'
    elif x >= 201701 and x < 201801:
        return '2017'
    elif x >= 201801 and x < 201901:
        return '2018'
    elif x > 201901 and x < 202001:
        return '2019'
    else:
        return '2020'
year_month_group(201509)
```

```python
df['yearmonth_by_year'] = df['yearmonth'].apply(year_month_group)
df
```

```python
df.AmountInUSD.value_counts()
```

```python
def Amount_In_USD(i):
    if i >= -999.0 and i < 10000000.0:
        return 'Group 1'
    elif i >= 10000000.0 and i < 20000000.0:
        return 'Group 2'
    elif i >= 20000000.0 and i < 30000000.0:
        return 'Group 3'
    elif i >= 30000000.0 and i < 40000000.0:
        return 'Group 4'
    else:
        return 'Outlier'
Amount_In_USD(700000.0)
```

```python
df['Amount_In_USD'] = df['AmountInUSD'].apply(Amount_In_USD)
df
```

```python
# 6. present the data above in a bar graph
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,30))
sns.countplot(x = 'Amount_In_USD', data=df)
plt.title("Number of funding per month")
plt.show()
```
